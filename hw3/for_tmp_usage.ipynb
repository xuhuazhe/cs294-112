```python
with open('./train_ddqn.log') as f_ddqn:
    count = 0
    iteration = []
    reward = []
    for content in f_ddqn.readlines():
        #print content.split()
        content = content.split()
        iteration.append(content[0])
        reward.append(content[1])
        count += 1
        
```
